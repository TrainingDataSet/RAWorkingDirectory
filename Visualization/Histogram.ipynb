```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("dataseries_GVC2007_2023.xlsx", sheet_name="2010_2019_Corr")
df
```

```python
# OECD국가와 Non-OECD국가 분리
oecd = df[df["dum_country_s_oecd"] == 1]
non_oecd = df[df["dum_country_s_oecd"] == 0]
```

```python
plt.figure(figsize=(8, 5))  # 그래프 크기 설정

# OECD 확률밀도함수
sns.kdeplot(
    oecd["Within_country_corr_CO2"], 
    linewidth=2,
    label="OECD", 
)

# Non-OECD 확률밀도함수
sns.kdeplot(
    non_oecd["Within_country_corr_CO2"], 
    linewidth=2,
    label="Non-OECD",
)

plt.title("Corr. betwen GDP and CO2 emission", fontsize=15)  # 제목 설정
plt.xlim(-1, 1)  # X값 절삭
plt.legend()     # 범례 표시
plt.xlabel("")   # X축 이름 설정

plt.xticks(np.arange(-1, 1.1, 0.5))  # X축 간격 조정
plt.yticks(np.arange(0, 1.1, 0.2))   # Y축 간격 조정

# X축과 Y축만 표시
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# X축과 Y축 크기 설정
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

plt.savefig("histogram_2.png", dpi=300)  # 파일 저장 (고해상도)
plt.show()
```
